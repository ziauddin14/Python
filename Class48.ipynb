```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\User\OneDrive\Desktop\Python-master\Python-master\Superstore.csv", encoding='latin1')

```

```python
sales = df['Sales'].values
profit = df['Profit'].values
discount = df['Discount'].values
quantity = df['Quantity'].values
```

```python
#find total sales using numpy
total_sales = np.sum(sales)
print("Total Sales:", total_sales)
```

```python
#find average profit using numpy
average_profit = np.mean(profit)
print("Average Profit:", average_profit)
```

```python
#find min and max sales using numpy
min_sales = np.min(sales)
max_sales = np.max(sales)
print("Minimum Sales:", min_sales)
print("Maximum Sales:", max_sales)
```

```python
#find total sales where profit is negative (losses)
loss_sales = np.sum(sales[profit < 0])
print("Total Sales with Losses:", loss_sales)
```

```python
#how many orders have a discount greater than 20%
high_discount_orders = np.sum(discount > 0.2)
print("Number of Orders with Discount > 20%:", high_discount_orders)
```

```python
# Find avg quantity sold where discount = 0
avg_quantity_no_discount = np.mean(quantity[discount == 0])
print("Average Quantity Sold with No Discount:", avg_quantity_no_discount)
```

```python
#increase all sales by 15%
increased_sales = sales * 1.15
print("Sales after 15% Increase:", increased_sales[:10])
```

```python
#create a new series called profit_flag (value: 1 = profit, 0 = loss)
profit_flag = np.where(profit > 0, 1, 0)
print("Profit Flag (1=Profit, 0=Loss):", profit_flag)
```
